```python
import scipy.stats as sc_s
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

wartosc = [1, 2, 3, 4, 5, 6]
prawdopodo = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6]
data = sc_s.rv_discrete(name='data', values=(wartosc, prawdopodo))
print("Mean {}".format(data.mean()))
print("Median {}".format(data.median()))
```

```python
bernoulli = sc_s.bernoulli.rvs(0.6, size=100)
binomial = sc_s.binom.rvs(6, 0.6, size=100)
poisson = sc_s.poisson.rvs(0.6, size=100)

print(bernoulli)
print(binomial)
print(poisson)
```

```python
fig, axBer = plt.subplots()
fig, axBin = plt.subplots()
fig, axPoi = plt.subplots()

sns.distplot(bernoulli, ax=axBer, color='Red')
sns.distplot(binomial, ax=axBin, color='Green')
sns.distplot(poisson, ax=axPoi, color='Blue')
```
